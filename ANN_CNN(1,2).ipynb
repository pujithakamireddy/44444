```python
//ANN with CIFAR-10
import tensorflow as tf
from tensorflow.keras import datasets,layers,models
import matplotlib.pyplot as plt
import numpy as np
(x_train,y_train),(x_test,y_test)=datasets.cifar10.load_data()
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
y_train[:50]
```

```python
clases=["airplane","automobile","bird","cat","dog","deer","frog","horse","ship"]
```

```python
def plot_sample(x,y,index):
  plt.figure(figsize=(15,1))
  plt.imshow(x[index])
  plt.xlabel(clases[y[index]])
```

```python
plot_sample
```

```python
y_train=y_train.reshape(-1,)
```

```python
y_test=y_test.reshape(-1,)
```

```python
plot_sample(x_train,y_train,10)
```

```python
x_train=x_train/255.0
```

```python
x_test=x_test/255.0
```

```python
ann=models.Sequential([layers.Flatten(input_shape=(32,32,3))])
layers.Dense(3000,activation='relu'),layers.Dense(1000,activation='relu'),layers.Dense(10,activation='relu')
```

```python
ann.compile(optimizer='SGD',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train,epochs=10)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
y_predict=ann.predict(x_test)
y_predict_clases=[np.argmax(element)for element in y_predict]
print("classification_report",classification_report(y_test,y_predict_clases))
```

```python
import tensorflow as tf
from tensorflow.keras import datasets,layers,models
import matplotlib.pyplot as plt
import numpy as np
(x_train,y_train),(x_test,y_test)=datasets.mnist.load_data()
```

```python
clases=["1","2","3","4","4","5"]
```

```python
def plot_sample(x,y,index):
  plt.figure(figsize=(15,3))
  plt.imshow(x[index])
  plt.xlabel(clases[y[index]])
```

```python
plot_sample(x_train,y_train,5)
```

```python
x_train=x_train/255.0
```

```python
ann=models.Sequential([layers.Flatten(input_shape=(28,28,3))])
layers.Dense(3000,activation='relu'),layers.Dense(1000,activation='relu'),layers.Dense(10,activation='relu')
```
